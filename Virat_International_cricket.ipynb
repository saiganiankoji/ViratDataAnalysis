```python
cities = ['Bangalore', 'Nagpur', 'Chennai', 'Indore', 'Visakhapatnam', 'Mumbai', 'Hyderabad',
          'Kolkata', 'Delhi', 'Rajkot', 'Pune', 'Guwahati', 'Jaipur', 'Dharamshala', 'Ranchi',
          'Mohali', 'Kanpur','Cuttack','Ahmedabad','Vadodara','Ranchi','Greater Noida','Guwahati','Trivandrum','Dehradun','Lucknow','Surat'
         ,'Raipur']

# Adding a new column 'H/A' based on 'Ground' column values
virat_df['H/A'] = virat_df['Ground'].apply(lambda x: 'Home' if any(city.lower() in x.lower() for city in cities) else 'Away')

# This code checks if any city from the 'cities' list is present in each 'Ground' value.
# If it finds a match (case insensitive), it assigns 'Home'; otherwise, it assigns 'Away' to the 'H/A' column.
virat_df
```

## Average Runs Against Each Team in Different Formats

```python
def calculate_average_runs(data_frame):
    return data_frame.groupby(['Opposition', 'Format'])['Runs'].mean().unstack()

# Calculate average runs
average_runs = calculate_average_runs(virat_df)
average_runs.replace(0, float('nan'), inplace=True)
average_runs.dropna(inplace=True)

# Plot the results
plt.figure(figsize=(12, 6))
bars=average_runs.plot(kind='bar', width=0.8, colormap='viridis')
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2), ha='center', va='bottom')
plt.title('Average Runs Against Each Team in Different Formats')
plt.xlabel('Opponent')
plt.ylabel('Average Runs')
plt.legend(title='Format')
plt.show()
```

```python
def calculate_average_runs(data_frame, match_format):
    format_data = data_frame[data_frame['Format'] == match_format]
    return format_data.groupby('Opposition')['Runs'].mean()

# Unique formats in the dataset
formats = virat_df['Format'].unique()

# Manually define a list of colors for each bar
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'pink', 'gray', 'cyan', 'yellow']

# Plot the results for each format
plt.figure(figsize=(20, 5 * len(formats)))

for i, match_format in enumerate(formats, 1):
    plt.subplot(len(formats), 1, i)
    
    # Calculate average runs for the current format
    average_runs = calculate_average_runs(virat_df, match_format)
    
    # Plot the results with different colors
    bars=plt.bar(average_runs.index, average_runs, color=colors[:len(average_runs)])
    for bar in bars.patches:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.5, round(yval, 2), ha='center', va='bottom')
    plt.title(f'Average Runs Against Each Team in {match_format}')
    plt.xlabel('Opponent')
    plt.ylabel('Average Runs')
    plt.legend([match_format], title='Format')

plt.tight_layout()
plt.show()
```
